```python
import matplotlib.pyplot as plt
import numpy as np

# Time axis: from 0 to 30
x = np.linspace(0, 30, 300)

# Initial phase: rapid growth (0~10)
y_init = 5 * np.log1p(x[x <= 10])  # log1p = log(1+x), for smooth curve

# Middle phase: plateau (10~20)
y_mid = np.full_like(x[(x > 10) & (x <= 20)], y_init[-1])

# Late phase: decline due to fatigue and stagnation (20~30)
y_late = y_mid[-1] - 0.3 * (x[x > 20] - 20) ** 1.5

# Combine all y values
y = np.concatenate([y_init, y_mid, y_late])
x_segments = np.concatenate([x[x <= 10], x[(x > 10) & (x <= 20)], x[x > 20]])

# Plot
plt.figure(figsize=(10, 5))
plt.plot(x_segments, y, label="Revenue Trend", color='navy')
plt.axvspan(0, 10, color='green', alpha=0.1, label='Initial Phase')
plt.axvspan(10, 20, color='orange', alpha=0.1, label='Middle Phase')
plt.axvspan(20, 30, color='red', alpha=0.1, label='Late Phase')

plt.title("Revenue Growth Curve (Initial - Middle - Late)")
plt.xlabel("Time")
plt.ylabel("Revenue")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 시간 축
x = np.linspace(0, 30, 300)

# 초기 (0~10): 빠른 성장 (exponential-like)
y_init = 1.5 * (np.exp(0.2 * x[x <= 10]) - 1)

# 중기 (10~20): 불만과 리스크 → 성장 둔화
y_mid = y_init[-1] + 2 * np.tanh((x[(x > 10) & (x <= 20)] - 10) / 2)

# 후기 (20~30): 신뢰 하락 → 성장이 멈춤
y_late = np.full_like(x[x > 20], y_mid[-1] - 1)

# 전체 y값 합치기
y = np.concatenate([y_init, y_mid, y_late])
x_segments = np.concatenate([x[x <= 10], x[(x > 10) & (x <= 20)], x[x > 20]])

# 그래프 그리기
plt.figure(figsize=(10, 5))
plt.plot(x_segments, y, label="Revenue Trend", color='darkred')
plt.axvspan(0, 10, color='limegreen', alpha=0.2, label='Initial Phase')
plt.axvspan(10, 20, color='gold', alpha=0.2, label='Middle Phase')
plt.axvspan(20, 30, color='lightcoral', alpha=0.2, label='Late Phase')

plt.title("Revenue Trend (Niche Growth → Legal Risk → Trust Stall)")
plt.xlabel("Time")
plt.ylabel("Revenue")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
